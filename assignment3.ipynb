1. Calculator using Python

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the Second number: "))
print("Press 'a' for Addition")
print("Press 's' for Subtraction")
print("Press 'm' for Multiplication")
print("Press 'd' for Division")
print("Press 'p' for Power")

opr = input("Enter any one character from a/s/m/d/p: ")

if opr == "a":
    print("The Addition is: ", num1+num2)
elif opr == "s": 
    print("The Subtraction is: ", num1 - num2)
elif opr == "m": 
    print("The Multiplication is: ", num1 * num2)
elif opr == "d": 
    print("The Division is: ", num1 / num2)
elif opr == "p":
    print("The Power of numbers is: ", num1 ** num2)
else:
    print("Enter the valid Character for Operation!")
```

2. Check Numeric Value in a list using loop

```python
numList = [23, 4, 52, 4,"shayan", "hdkhaksd", 60]
count = 0
for i in numList:
    if type(i) == int or type(i) == float:
        count+=1
        print(i)

if count > 0:
    print(f"{count} numeric items exits in list!")
else:
    print("No numeric item exits in a list!")
```

3. Add a new key inside a Dictionary

```python
my_dict = {"name": "Ali", "age": 20}
print("Original Dictionary:", my_dict)

# Adding a new key "city" with the value "Karachi"
my_dict["city"] = "Karachi"

print("Updated Dictionary:", my_dict)
```

4. Sum all numeric values inside a dictionary

```python
my_dict = {'item1': 100, 'item2': 'apple', 'item3': 250, 'item4': 50}
total_sum = 0

for value in my_dict.values():
    if type(value) == int or type(value) == float:
        total_sum = total_sum + value

print("Sum of all numeric items:", total_sum)
```

5. Find Duplicate values inside a list

```python
list = [2,5,3,4,2,"shayan", "ali", "shayan", "ali", 5]
dup = []
for i in range(len(list)):
    for j in range(i+1, len(list)):
        if list[i] == list[j]:
          dup.append(list[i])

print("The duplicated items in a list is: ", dup)
```
